```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

```python
url="https://trucksuvidha.com/"
```

```python
r= requests.get(url)
```

```python
soup = BeautifulSoup(r.text,"lxml")
```

```python
table = soup.find("table",class_ = "table table-hover table-condensed even tbl-border-colapse")
print(table)
```

```python
headers = table.find_all("th")
print(headers)
```

```python
titles = []
for i in headers:
    title = i.text
    titles.append(title)

print(titles)
```

```python
import pandas as pd
df = pd.DataFrame(columns=titles)
print(df)
```

```python
rows = table.find_all("tr")
#print(rows)
```

```python
#for i in rows[1:]:
    #print(i.text)
    
```

```python
for i in rows[1:]:
    data = i.find_all("td")
    #print(data)
    row = [tr.text for tr in data]
    #print(row)
    l= len(df)
    df.loc[l]= row

print(df)
    
```

```python
def clean_dataframe(df):
    return df.applymap(lambda x: x.strip() if isinstance(x, str) else x)
```

```python

cleaned_df = clean_dataframe(df)
print(cleaned_df)
```

```python
 #Drop the last two columns
cleaned_df = cleaned_df.iloc[:, :-2]

# Save the cleaned data to a CSV file
cleaned_df.to_csv("cleaned_data.csv", index=False)

print(cleaned_df)
```
