```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/trucksuvidha_data.csv")
```

```python
for column in df.columns:
    if pd.api.types.is_numeric_dtype(df[column]): # Check if the column is numeric
        plt.figure(figsize=(8, 6))  # Adjust figure size as needed
        sns.histplot(df[column], kde=True)  # Create histogram with kernel density estimation
        plt.title(f'Distribution of {column}')
        plt.xlabel(column)
        plt.ylabel('Frequency')
        plt.show()
    else:  # For categorical columns
        plt.figure(figsize=(8, 6))
        df[column].value_counts().plot(kind='bar')
        plt.title(f'Distribution of {column}')
        plt.xlabel(column)
        plt.ylabel('Frequency')
        plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
        plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

for column in df.columns:
    if pd.api.types.is_numeric_dtype(df[column]):
        plt.figure(figsize=(8, 6))

        # Choose a suitable plot based on data distribution
        if len(df[column].unique()) > 10:  # If many unique values, use histogram or KDE
            sns.histplot(df[column], kde=True)
        else:  # If few unique values, use boxplot or violinplot
            sns.boxplot(y=df[column])

        plt.title(f'Distribution of {column}')
        plt.xlabel(column)
        plt.ylabel('Frequency / Values')  # Adjust y-label as needed
        plt.show()

    elif pd.api.types.is_categorical_dtype(df[column]) or pd.api.types.is_object_dtype(df[column]):
        plt.figure(figsize=(8, 6))

        # Choose a suitable plot for categorical data
        value_counts = df[column].value_counts()
        if len(value_counts) > 10:  # If many categories, use bar plot with limited top categories
            value_counts[:10].plot(kind='bar')  # Show top 10 categories
            plt.xticks(rotation=45, ha='right')
        else:  # If few categories, use pie chart or countplot
            # sns.countplot(x=df[column])  # Use countplot for fewer categories
            df[column].value_counts().plot(kind='pie', autopct='%1.1f%%') # Pie chart

        plt.title(f'Distribution of {column}')
        plt.xlabel(column)
        plt.ylabel('Frequency / Proportion')  # Adjust y-label as needed
        plt.show()
```
